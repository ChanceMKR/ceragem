- 저는 html(LDA 시각화 파일) 열어서 각 Action에 사용한 topic 누르고, 나오는 키워드들 중 상단부터 보면서 넣어봤습니다...

#### 헷갈릴 수 있으니 본인 담당 액션정리
- [topic1, 7] / action1 초보식집사
- [topic2] / action2 채소재배
- [topic5] / action3 인테리어 소품
- [topic3] / action4 베란다

```python
#LDA 키워드 #장표순서
action1=['초보', '키우', '화분','분갈이','뿌리','마사토','자르','마르','물주','어렵','처음']
action2= ['텃밭','상추','모종','씨앗','채소','과일','고추','방울토마토','수확','대파','발아','배추','작물','열매','농사','부추','호박','오이','토마토']
action3= ['플랜테리어','식물 인테리어','인테리어','소품','신혼집','거실','행잉','조명','공기정화','필름','선물','정화','이쁘','예쁘', '분위기','꾸미','화이트','초록', '어울리','귀엽','다육','이름']
action4= ['옮기','창문','소리','베란다','냄새','벌레','방충망','들어오','테라스']


# 사전 만들기
name=[action1, action2, action3, action4]

col_num=len(name)+1

dic={}
for n in range(1,col_num):
    dic[f'action_{n}'] = name[n-1]
dic
```

# 키워드 카운트

```python
sent_list = sent.review.values.ravel().tolist()

def calculate_keyword_frequency(data, keywords, topic_num):
    list_=[]
    for row in data:

        # total 초기화
        total=0

        # 키워드별 빈도수를 저장할 딕셔너리 초기화
        frequencies = {keyword: 0 for keyword in keywords}

        for keyword in keywords:
            if keyword in row:
                frequencies[keyword] += 1

        # value값 다 더하기
        for value in frequencies.values():
            total+=value

        list_.append(total)

    return pd.DataFrame({topic_num: list_})
```

```python
to=[]
for i in range(1, col_num):
    result = calculate_keyword_frequency(sent_list, dic[f'action_{i}'], f'action_{i}')
    to.append(pd.DataFrame(result))
to.append(sent)

total=pd.concat(to, axis=1)



#-------------------------자신의 action 번호에 맞게 변경 필요---------------------------

# dictionary점수 모든 열이 0인 값 vs. 0이 아닌값을 분리
zero_rows = total[(total['action_1'] == 0)&(total['action_2'] == 0) &
                  (total['action_3'] == 0)& (total['action_4'] == 0)]

non_zero_rows = total[~((total['action_1'] == 0)&(total['action_2'] == 0) &
                        (total['action_3'] == 0)& (total['action_4'] == 0))]


#-------------------------------------------------------------



# 키워드 빈도 합이 가장 큰 값을 가지는 action을 최종 action으로 할당
col=[f'action_{i}' for i in range(1, col_num)]
max_column = non_zero_rows[col].idxmax(axis=1)

non_zero_rows['final_action_num']=[_[-1:] for _ in max_column]

print(non_zero_rows.final_action_num.value_counts())

display(non_zero_rows)
display(zero_rows)

print('재할당 후 reviewer 수: ',len(non_zero_rows.reviewer_id.unique()), '\n원래 전체 reviewer 수: ', len(df))
```
